# Load Necessary Libraries

```python
#Data Visualization in Python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
plt.plot([1,2,3], [2,4,6])
```

```python
# To clean up, move arrays into variables

x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize graph
# Second parameter specifies how many pixels it is
# Smaller values give you more pixelated graphs
plt.figure(figsize = (5,3), dpi = 300 )


# Passing in a label
# Altering graph
#plt.plot(x,y, label = '2x', color = 'red', linewidth = 2,
         #marker = '.', markersize = 10, markeredgecolor = 'blue',
         #linestyle = '--')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'r.--', label = '2x')


# Line Number Two
x2 = np.arange(0,4.5,0.5)
print(x2)
#plt.plot(x2, x2**2, 'g', label = 'x^2')

#Plotting projections
plt.plot(x2[:6], x2[:6]**2, 'g', label = 'x^2')
plt.plot(x2[5:], x2[5:]**2, 'g--')

# Add titles
plt.title('Our First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Change the tick marks on the axis
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

# Adding a Legend
plt.legend() 

# Saving a graph
plt.savefig('mygraph.png', dpi = 300)

# In order to remove the text at the top of the graph 
plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

# Adding symbols to bars
#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')

# Another way of adding patterns, different notation
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
# Resize the graph
plt.figure(figsize = (6,4))

plt.show()
```
